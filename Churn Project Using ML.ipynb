```python
x_test_transformed = pd.concat([x_test_cat_ohe,x_test_num_rescale],axis=1)
x_test_transformed
```

# Importing sklearn models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
reg_model = [LogisticRegression(),
             DecisionTreeClassifier(),
             RandomForestClassifier(n_estimators=200),
             SVC(),
             GradientBoostingClassifier()]

score_card = pd.DataFrame(columns=["Algorithms", "Accuracy"])

result = []
alg = []

for model in reg_model:
    model_name = model.__class__.__name__
    model.fit(x_train_transformed, y_train)
    y_test_predicted = model.predict(x_test_transformed)
    accuracy = accuracy_score(y_test, y_test_predicted)
    result.append(accuracy)
    alg.append(model_name)

score_card["Algorithms"] = alg
score_card["Accuracy"] = result

score_card
```
